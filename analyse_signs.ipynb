```
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```
train= pd.read_pickle('ML_road_sign_recognition/data/train.p')
```

```
X_train,y_train=train['features'], train['labels']
```

```
X_train.shape,y_train.shape
```

```
plt.imshow(X_train[0])
```

```
signs= pd.read_csv('ML_road_sign_recognition/data/signnames.csv')
```

```
signs.head()
```

```
dict_signs=signs.to_dict()['b']
dict_signs[30]
```

```
plt.imshow(X_train[y_train==0][0])
```

```
for id_sign in dict_signs.keys():
  given_signs=X_train[y_train==id_sign]

  for i in range(9):
    plt.subplot('19{0}'.format(i+1))
    plt.imshow(given_signs[i])
    plt.axis('off')
  
  plt.tight_layout()
  plt.show()
```

```
cnt=Counter(y_train).most_common()
```

```
id_label,occurance=zip(*cnt)
```

```
ids=range(len(id_label))

plt.figure(figsize=(18,8))
plt.bar(id_label,occurance)

labels=[dict_signs[id_label[id_]] for id_ in id_label]
plt.xticks(ids,labels, rotation='vertical')

plt.show()
```
